#### There are 120 channels that have downloads, there are 36 channels without a single download. Figure out the clicks avg of downloaded channels vs not downloaded channels

```python
print(f'Average Clicks for Channels with zero downloads: {no_channels.count_clicks.mean().round(5)}')
print(f'Average Clicks for Channels with at least one download: {yes_channels.count_clicks.mean().round(5)}\n')

print(f'Sample Standard Deviation for channels w/ 0 downloads: {no_channels.count_clicks.std().round(5)}')
print(f'Sample SD for channels w/ at least one download: {yes_channels.count_clicks.std().round(5)}')
```

```python
print(f"Conducting Welch's T-test, due to the sample size being two different lengths\n")
print(f"Hnull: There is no difference between channels' average clicks with downloads and channels' average clicks w/o downloads. AKA mean(downloads) = mean(no_downloads)")
print(f"Halternative: There is a difference between channels' average clicks w/ downloads & channels' avg clicks w/o downloads. AKA mean(downloads) != mean(no_downloads)")

ttest,pval = stats.ttest_ind(no_channels['count_clicks'], yes_channels['count_clicks'], equal_var=False)
print("p-value = ",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we fail to reject null hypothesis")
```

```python
outliers=[]
def detect_outlier(data_1):
    
    threshold=3
    mean_13 = devices.device.mean()
    std_13 = devices.device.std()
    
    
    for y in data_1:
        z_score= (y - mean_13)/std_13 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```

```python
outlier_datapoints = detect_outlier(devices.device)
print(outlier_datapoints)
```

```python
dev = train2.groupby(['device']).count().sort_values(by=['click_time'],ascending=False).reset_index()
dev.drop(columns=['ip','app','os','channel'],inplace=True)
dev = dev.loc[dev.attributed_time!=0].reset_index()
dev
```

### Conducting a chi-square test for devices and their conversion rate
- Hnull: Device conversion rates are uniformly distributed
- Halternative: Device conversion rates are not uniformly distributed

#### Xcritical w/ 8 dof at alpha 0.05 = 15.507

```python
dof = 9-1
xbar = ((dev.conversion_rate.sum())/9)
print(f'Degrees of Freedom: {dof}\n Expected: {xbar}')
```

```python
chi_dev = dev.copy()
chi_dev.drop(columns=['count_clicks','count_downloads'],inplace=True)
chi_dev['expected'] = ((xbar))
chi_dev.rename(columns={'conversion_rate':'observed'},inplace=True)
```

```python
chi_dev
chi_dev['O-E'] = chi_dev.observed - chi_dev.expected
chi_dev['(O-E)Squared'] = (chi_dev.observed - chi_dev.expected)**2
chi_dev['(O-E)Squared/E'] = (chi_dev['(O-E)Squared'])/chi_dev.expected
chi_dev
```

```python
xchi_square = chi_dev['(O-E)Squared/E'].sum()
```

```python
if xchi_square < 15.507:
  print("We fail to reject the null hypothesis.")
else:
  print("We reject the null hypothesis.")
```
